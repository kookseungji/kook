```python
import requests
from bs4 import BeautifulSoup
url='https://finance.naver.com/sise/sise_index_day.nhn?'

date_list=[]
price_list=[]
for page in range(557):
    params={
    'code':'KPI200',
    'page':page} 
    resp=requests.get(url,params)
    soup=BeautifulSoup(resp.content)
    div=soup.find_all('tr')
    for i in div:    
            try:
                date=i.find('td',class_='date').text.replace('.','')
                price=i.find('td',class_='number_1').text
                print(date)
                print(price)
                date_list.append(date)
                price_list.append(float(price))
            except:
                pass
```

```python
import pandas as pd
kospi_dict={}
kospi_dict['date']=date_list
kospi_dict['kospi']=price_list
kospi=pd.DataFrame(kospi_dict)
```

```python
import json
import pandas as pd
from urllib.request import urlopen
import requests
date_list=[]
S_P500_list=[]
url='https://finance.naver.com/world/worldDayListJson.nhn?symbol=SPI@SPX&fdtc=0'

for page in range(441):
    params={'page':page}
    raw=requests.get(url,params).json()
    for i in raw:
        date=i['xymd'] 
        S_P500=i['clos']
        date_list.append(date)
        S_P500_list.append(S_P500)
```

```python
snp_dict={}
snp_dict['date']=date_list
snp_dict['S_P500']=S_P500_list
```

```python
snp=pd.DataFrame(snp_dict)
```

```python
df=pd.merge( snp,kospi,how='inner',on='date')
```

```python
df=df.sort_values(by='date',ascending=True)
df=df.reset_index(drop=True)
```

```python
df['date']=df['date'].apply(lambda x : x[:4]+'-'+x[4:6]+'-'+x[6:])
```

```python
df=df.set_index('date')
```

```python
df.index=pd.to_datetime(df.index)
df
```
