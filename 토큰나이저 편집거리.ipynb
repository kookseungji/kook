```python
import nltk
nltk.download('punkt')
from nltk.tokenize import LineTokenizer,SpaceTokenizer, TweetTokenizer
from nltk import word_tokenize
```

```python
ITokenizer=LineTokenizer()
print('Line Tokenizer 출력:',ITokenizer.tokenize('My name is lalalat'))#개행 기준 한줄 인식

raw='By 11 o\'clock on Sunday, the doctor shall open the dispensary'
sTokenizer =SpaceTokenizer()
print('Space Tokenizer 출력:',sTokenizer.tokenize(raw))

print('Word Tokenizer 출력:',word_tokenize(raw)) #,도 나옴

tTokenizer=TweetTokenizer()
tTokenizer.tokenize('This is a cool #dum :)')#:)도 나옴
```

```python
from nltk import PorterStemmer, LancasterStemmer, word_tokenize
raw='MY name is lalalaldlsdf, supervisor segmentation paging'
tokens=word_tokenize(raw)
print(tokens)
```

```python
porter=PorterStemmer()
pStems =[porter.stem(t) for t in tokens]
print(pStems)
```

```python
lancaster=LancasterStemmer()
Stems=[lancaster.stem(t) for t in tokens]
print(Stems)
```

```python
from nltk.metrics.distance import edit_distance
```

```python
def edit(str1,str2):
    m=len(str1)+1
    n=len(str2)+1
    
    table={}
    for i in range(m): table[i,0]=i
    for j in range(n): table[0,j]=j # 테이블 가장자리 셋팅
    for i in range(1,m):
        for j in range(1,n):
            cost=0 if str1[i-1]==str2[j-1] else 1
            table[i,j]=min(table[i,j-1]+1,table[i-1,j]+1 ,table[i-1,j-1]+cost)
            #테이블 내부 채우기
    return table[i,j]
```
