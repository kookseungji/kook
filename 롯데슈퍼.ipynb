```python
def mart_data(text):
    # -*- coding: utf-8 -*-
    with open(text, 'rt', encoding='utf-8') as f:
        lines  = f.readlines(67000000)

    import pandas as pd
    dt=pd.DataFrame(lines)
    dt=dt[0].str.split('\t', expand=True)    
    dt.columns=['매장코드','매장이름','판매일','포스번호','영수증번호','판매시간','상품납품코드','상품슈퍼코드','상품명','판매수량','중량','판매금액','순판매금액']
    import re
    dt['순판매금액']=dt['순판매금액'].apply(lambda x:re.sub('\n','',x))
    dt['순판매금액']=dt['순판매금액'].apply(lambda x:int(x))
    dt['판매수량']=dt['판매수량'].apply(lambda x:int(x))
    dt['판매일']=pd.to_datetime(dt['판매일'])
    return dt
```

```python
def table(dt):
    dt0=dt.groupby(['상품명','판매일']).sum()['판매수량'].unstack().fillna(0)
    return dt0
```
