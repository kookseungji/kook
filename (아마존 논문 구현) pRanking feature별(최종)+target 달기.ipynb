```python
total=pd.read_csv('total.csv')
```

```python
import numpy as np

def test(text, n):
    if text == 'NEG':
        return 1/n
    return n

def seonmin(idx):
    if idx in idx_list:
        return data['n'][idx+1]/data['n'][idx]
    else:
        return data['n'][idx]
    
def main(feature):
    data=total[total['FeatureName']==feature]
    data=data.groupby(['base제품명','SS/CS','target','Label','title','asin'],as_index=False).count()
    data=data.rename(columns={'FeatureName':'n'})
    data.reset_index(inplace=True)
    data['n']=data.apply(lambda x : test(x['Label'], x['n']), axis=1)
    
    이거=data.drop(['n','index'],axis=1).groupby(['base제품명','SS/CS','target','title','asin'],as_index=False).count()
    이거=이거[이거['Label']!=1].reset_index(drop=True)
    df=pd.merge(이거,data,on=['base제품명','SS/CS','target','title','asin'],how='left')
    df.groupby(['base제품명','SS/CS','target','title','asin','Label_x'],as_index=False)
    df.set_index('index',inplace=True)
    idx_list=list(df.index.values[np.arange(0,len(df)-1,2)])
    data['n']=data['index'].apply(lambda x:seonmin(x))
    
    nodes=set(data['base제품명'])
    edges=[]
    for i in range(len(data)):
        edges.append((data['base제품명'][i],data['target'][i],data['n'][i]))
    graph=nx.DiGraph()
    graph.add_nodes_from(nodes)
    graph.add_weighted_edges_from(edges)
    graph=nx.pagerank(graph)
    
    data_feature=pd.DataFrame(graph, index=range(1)).T.reset_index().rename(columns = {'index':'base제품명', 0 : 'score'})
    data_feature.sort_values('score',ascending=False,inplace=True)
    data_2=total[total['FeatureName']==feature]
    
    title_list = data_2.drop(['target','SS/CS','FeatureName','asin','Label'], axis = 1).reset_index(drop=True)
    title_list = title_list.drop_duplicates(['base제품명','title'], keep='first').reset_index(drop=True)
    feature_ranking = data_feature.merge(title_list, on='base제품명', how='left')
    feature_ranking = feature_ranking.drop_duplicates('base제품명', keep='first').reset_index(drop=True)
    last_10 = feature_ranking[:10]
    last_10.to_csv('VOC_' + feature + '_rank_TOP10.csv', encoding='EUC-KR', index=0)
    return last_10
```
