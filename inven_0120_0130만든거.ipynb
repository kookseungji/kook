```python
# -*- coding: utf-8 -*-
with open('inven_0130.dat', 'rt', encoding='utf-16') as f:
    lines  = f.readlines()
```

```python
import pandas as pd
dt=pd.DataFrame(lines)
dt=dt[0].str.split('\t', expand=True)
```

```python
dt.columns=['timestamp','점포코드','점포명','상품코드','item_id','demand(first)','demand(last)', '단위/중량']
```

```python
import datetime
dt['timestamp']=dt['timestamp'].apply(lambda x :datetime.datetime.strptime(str(x),'%Y%m%d'))
```

```python
dt.to_csv('inven_0130.csv',encoding='utf-8-sig')
```

```python
import re
dt['단위/중량']=dt['단위/중량'].apply(lambda x:re.sub("(\\n)", "",str(x)))
```
